```python
rootdir0 = '../numpy_data/C0A'
rootdir1 = '../numpy_data/C1A'
rootdir2 = '../numpy_data/C2A'
rootdir3 = '../numpy_data/C3A'
rootdir4 = '../numpy_data/C4A'
rootdir5 = '../numpy_data/C5A'
rootdir6 = '../numpy_data/C6A'

list_0 = os.listdir(rootdir0) #列出文件夹下所有的目录与文件
list_1 = os.listdir(rootdir1)
list_2 = os.listdir(rootdir2)
list_3 = os.listdir(rootdir3)
list_4 = os.listdir(rootdir4)
list_5 = os.listdir(rootdir5)
list_6 = os.listdir(rootdir6)


lab0 = np.array([0])
lab1 = np.array([1])
lab2 = np.array([2])
lab3 = np.array([3])
lab4 = np.array([4])
lab5 = np.array([5])
lab6 = np.array([6])

num_train = 0 
num_test = 0
```

```python
snr_num = -4
snr_str = "-4_db"
```

```python
def wgn(x, snr):
    
    snr = 10**(snr/10.0)
    xpower = np.sum(x**2)/len(x)
    npower = xpower / snr
    
    return np.random.randn(len(x)) * np.sqrt(npower)


def add_noise(data):
    
    rand_data = wgn(data, snr_num)
    
    for id in range(len(data)):
        data[id] = data[id] + rand_data[id]
    
    return data
```

```python
detail_num = 0

for id_0 in range(0,len(list_0)):
    
    path = os.path.join(rootdir0,list_0[id_0])
    
    if os.path.isfile(path):
        
        print(path)
        
        data_npy = np.load(path)
        data_npy = add_noise(data_npy)
        
        train_data = data_npy[:384000]
        test_data = data_npy[384000:]
                       
        train_len = (len(train_data)-4096)/2048 + 1
        test_len = (len(test_data)-4096)/2048 + 1
        
        train_len = int(train_len)
        test_len = int(test_len)
        
        for sub_id in range(0, train_len):
            
            sub_train_data = train_data[sub_id*2048:sub_id*2048+4096]
            
            
            str_num_train = str(num_train)
            np.save("../IB_data/noise_data/"+snr_str+"/train_data/"+ str_num_train + "_train.npy", sub_train_data)
            np.save("../IB_data/noise_data/"+snr_str+"/train_lab/"+ str_num_train + "_lab.npy", lab0)    
            
            num_train += 1
        
        for sub_id in range(0, test_len):
            
            sub_test_data = test_data[sub_id*2048:sub_id*2048+4096]
            
            
            str_num_test = str(num_test)
            np.save("../IB_data/noise_data/"+snr_str+"/test_data/"+ str_num_test + "_test.npy", sub_test_data)
            np.save("../IB_data/noise_data/"+snr_str+"/test_lab/"+ str_num_test + "_lab.npy", lab0) 
            
            str_detail_num = str(detail_num)
            np.save("../IB_data/noise_data/"+snr_str+"/test_detail/0/"+ str_detail_num + "_test.npy", sub_test_data)
            np.save("../IB_data/noise_data/"+snr_str+"/test_lab_detail/0/"+ str_detail_num + "_lab.npy", lab0)       
            detail_num += 1
            
            num_test += 1
```
